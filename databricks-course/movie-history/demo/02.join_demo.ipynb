### Spark Join

#### **INNER JOIN**

```python
%run "configuration"
```

```python
df_movies = spark.read.parquet(f"{silver_folder_path}/movies").filter("year_release_date=2007")
display(df_movies)
```

```python
df_production_country = spark.read.parquet(f"{silver_folder_path}/production_countries")
display(df_production_country)
```

```python
df_country = spark.read.parquet(f"{silver_folder_path}/countries")
display(df_country)
```

```python
df_movie_production_country = df_movies.join(df_production_country,df_movies.movie_id==df_production_country.movie_id).select(df_movies.title,df_movies.budget,df_production_country.country_id)
```

```python
df_movie_country = df_movie_production_country.join(df_country,df_movie_production_country.country_id==df_country.country_id,"inner") \
                                              .select(df_movie_production_country.title,df_movie_production_country.budget,df_country.country_name)
display(df_movie_country)
```

#### **OUTER JOIN**

#### Left Outer Join

```python
df_movie_production_country = df_movies.join(df_production_country,df_movies.movie_id==df_production_country.movie_id,"left")\
                                       .select(df_movies.title,df_movies.budget,df_production_country.country_id)
display(df_movie_production_country)
```

```python
df_movie_country = df_movie_production_country.join(df_country,df_movie_production_country.country_id==df_country.country_id,"left") \
                                              .select(df_movie_production_country.title,df_movie_production_country.budget,df_country.country_name)
display(df_movie_country)
```

#### Right Outer Join

```python
df_movie_production_country = df_movies.join(df_production_country,df_movies.movie_id==df_production_country.movie_id,"right")\
                                       .select(df_movies.title,df_movies.budget,df_production_country.country_id)
display(df_movie_production_country)
```

```python
df_movie_country = df_movie_production_country.join(df_country,df_movie_production_country.country_id==df_country.country_id,"right") \
                                              .select(df_movie_production_country.title,df_movie_production_country.budget,df_country.country_name)
display(df_movie_country)
```

#### Full Outer Join

```python
df_movie_production_country = df_movies.join(df_production_country,df_movies.movie_id==df_production_country.movie_id,"full")\
                                       .select(df_movies.title,df_movies.budget,df_production_country.country_id)
display(df_movie_production_country)
```

```python
df_movie_country = df_movie_production_country.join(df_country,df_movie_production_country.country_id==df_country.country_id,"full") \
                                              .select(df_movie_production_country.title,df_movie_production_country.budget,df_country.country_name)
display(df_movie_country)
```

#### Anti Join
